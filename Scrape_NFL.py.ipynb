#  National Football League stadiums Img and info

```python
table_url = "https://en.wikipedia.org/wiki/List_of_current_National_Football_League_stadiums"
browser.visit(table_url)
html = browser.html
soup = bs(html,"html.parser")
```

```python
#print(soup.prettify())
```

```python
# list of all stadiums fot the for loop
stadiums_strings = final_stadiums['Name'].values.tolist()
stadiums_strings
```

```python
# a for loop that builds a list of dics. in each one is a title and info for each stadium  
stadiums_info_urls = []

for stadium in stadiums_strings:
    
    stadiums_dict = {}
    
    browser.click_link_by_partial_text(stadium)
    html = browser.html
    soup = bs(html,"html.parser")
    
#     stadiums_dict["img_url: "] = browser.find_by_css('h1[class="site-headline"] img')
    
    stadiums_dict["title: "] = stadium
    
    stadiums_dict["info: "] = soup.find('p')
    
    stadiums_info_urls.append(stadiums_dict)
    
    #going back to the landing page
    browser.back()
```
